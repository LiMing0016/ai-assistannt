```python
import requests
import json

resp = requests.post(
    "http://127.0.0.1:8013/assistant/chat",
    json={
        "message": "请解释一下现在完成时和一般过去时的核心区别，并各给两个例句",
        "conversation_id": "jupyter-test-1",
        "provider": "openai",
        "model": "gpt-5-mini"
    },
    timeout=60,
)

print(resp.status_code)
print(json.dumps(resp.json(), ensure_ascii=False, indent=2))
```

```python
cases = [
    {
        "message": "请解释一下现在完成时和一般过去时的核心区别，并各给两个例句",
        "conversation_id": "jupyter-explain",
        "provider": "openai",
        "model": "gpt-5-mini",
    },
    {
        "message": "先把这句话翻译成英文：我昨天去了图书馆。然后解释为什么不能说 I go to library yesterday. 最后给我两道同类练习。",
        "conversation_id": "jupyter-multi",
        "provider": "openai",
        "model": "gpt-5-mini",
    },
    {
        "message": "顺便帮我做一个北京三日旅游计划",
        "conversation_id": "jupyter-boundary",
        "provider": "openai",
        "model": "gpt-5-mini",
    },
]

for case in cases:
    resp = requests.post(
        "http://127.0.0.1:8013/assistant/chat",
        json=case,
        timeout=60,
    )
    print("=" * 80)
    print(case["conversation_id"], resp.status_code)
    print(json.dumps(resp.json(), ensure_ascii=False, indent=2))
```

```python
import requests
import json

cases = [
    {
        "label": "primary",
        "payload": {
            "message": "我写的是 I go to library yesterday. 这样翻对吗？请帮我改一下并解释。",
            "conversation_id": "stage-compare-primary",
            "user_id": "u-primary",
            "provider": "openai",
            "model": "gpt-5-mini",
        },
    },
    {
        "label": "ielts",
        "payload": {
            "message": "我写的是 I go to library yesterday. 这样翻对吗？请帮我改一下并解释。",
            "conversation_id": "stage-compare-ielts",
            "user_id": "u-ielts",
            "provider": "openai",
            "model": "gpt-5-mini",
        },
    },
]

for case in cases:
    resp = requests.post(
        "http://127.0.0.1:8001/assistant/chat",
        json=case["payload"],
        timeout=60,
    )
    print("=" * 80)
    print(case["label"], resp.status_code)
    data = resp.json()
    print(json.dumps(data, ensure_ascii=False, indent=2))

print("\n--- reply ---\n")
print(data["reply"])
```
